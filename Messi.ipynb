1. Desarrolle un programa que permita realizar la suma/resta de matrices. El programa debe validar las condiciones necesarias para ejecutar la operación.

```python
# Función para crear una matriz solicitando los valores al usuario
def crear_matriz(filas, columnas):
    matriz = []  # Inicializamos una lista vacía para almacenar la matriz
    for i in range(filas):  # Iteramos sobre el número de filas
        fila = []  # Inicializamos una lista vacía para cada fila
        for j in range(columnas):  # Iteramos sobre el número de columnas
            # Solicitamos al usuario que ingrese el valor para la posición [i+1][j+1]
            valor = int(input(f"Ingrese el valor para la posición [{i+1}][{j+1}]: "))
            fila.append(valor)  # Agregamos el valor a la fila actual
        matriz.append(fila)  # Agregamos la fila completa a la matriz
    return matriz  # Devolvemos la matriz creada


# Función para imprimir una matriz de forma legible
def imprimir_matriz(matriz):
    for fila in matriz:  # Iteramos sobre cada fila de la matriz
        print(fila)  # Imprimimos la fila actual


# Función para sumar dos matrices
def sumar_matrices(matriz1, matriz2):
    filas = len(matriz1)  # Obtenemos el número de filas de la primera matriz
    columnas = len(matriz1[0])  # Obtenemos el número de columnas de la primera matriz
    # Creamos una matriz de resultado llena de ceros con las mismas dimensiones
    resultado = [[0 for _ in range(columnas)] for _ in range(filas)]

    # Iteramos sobre cada posición de la matriz
    for i in range(filas):
        for j in range(columnas):
            # Sumamos los valores correspondientes de ambas matrices
            resultado[i][j] = matriz1[i][j] + matriz2[i][j]

    return resultado  # Devolvemos la matriz resultante


# Función para restar dos matrices
def restar_matrices(matriz1, matriz2):
    filas = len(matriz1)  # Obtenemos el número de filas de la primera matriz
    columnas = len(matriz1[0])  # Obtenemos el número de columnas de la primera matriz
    # Creamos una matriz de resultado llena de ceros con las mismas dimensiones
    resultado = [[0 for _ in range(columnas)] for _ in range(filas)]

    # Iteramos sobre cada posición de la matriz
    for i in range(filas):
        for j in range(columnas):
            # Restamos los valores correspondientes de ambas matrices
            resultado[i][j] = matriz1[i][j] - matriz2[i][j]

    return resultado  # Devolvemos la matriz resultante


# Función principal del programa
def main():
    # Solicitar al usuario las dimensiones de las matrices
    filas = int(input("Ingrese el número de filas: "))
    columnas = int(input("Ingrese el número de columnas: "))

    # Crear la primera matriz
    print("\nCreando la primera matriz:")
    matriz1 = crear_matriz(filas, columnas)

    # Crear la segunda matriz
    print("\nCreando la segunda matriz:")
    matriz2 = crear_matriz(filas, columnas)

    # Validar que las matrices tengan las mismas dimensiones
    if len(matriz1) != len(matriz2) or len(matriz1[0]) != len(matriz2[0]):
        print("Las matrices no tienen las mismas dimensiones. No se puede realizar la operación.")
        return  # Terminar el programa si las dimensiones no coinciden

    # Mostrar las matrices ingresadas
    print("\nMatriz 1:")
    imprimir_matriz(matriz1)

    print("\nMatriz 2:")
    imprimir_matriz(matriz2)

    # Solicitar al usuario la operación que desea realizar
    operacion = input("\n¿Qué operación desea realizar? (suma/resta): ").lower()

    # Realizar la operación seleccionada
    if operacion == "suma":
        resultado = sumar_matrices(matriz1, matriz2)
        print("\nResultado de la suma:")
        imprimir_matriz(resultado)
    elif operacion == "resta":
        resultado = restar_matrices(matriz1, matriz2)
        print("\nResultado de la resta:")
        imprimir_matriz(resultado)
    else:
        print("Operación no válida.")  # Mensaje si la operación no es válida


# Punto de entrada del programa
if __name__ == "__main__":
    main()  # Llamar a la función principal para ejecutar el programa
```

2. Desarrolle un programa que permita realizar el producto de matrices. El programa debe validar las condiciones necesarias para ejecutar la operación.

```python
def ingresar_matriz(filas, columnas):
    matriz = []
    for i in range(filas):
        while True:
            fila = input(f"Ingrese la fila {i+1} (separada por espacios): ").strip().split()
            if len(fila) == columnas:
                try:
                    fila = [float(valor) for valor in fila]
                    matriz.append(fila)
                    break
                except ValueError:
                    print("Error: Ingrese solo números válidos.")
            else:
                print(f"Error: La fila debe tener {columnas} elementos.")
    return matriz

def multiplicar_matrices(matriz1, matriz2):
    filas_matriz1 = len(matriz1)
    columnas_matriz1 = len(matriz1[0])
    filas_matriz2 = len(matriz2)
    columnas_matriz2 = len(matriz2[0])

    # Verificar si el producto es posible
    if columnas_matriz1 != filas_matriz2:
        raise ValueError("No se puede multiplicar: el número de columnas de la primera matriz no coincide con el número de filas de la segunda matriz.")

    # Inicializar la matriz resultado con ceros
    resultado = [[0 for _ in range(columnas_matriz2)] for _ in range(filas_matriz1)]

    # Realizar la multiplicación
    for i in range(filas_matriz1):
        for j in range(columnas_matriz2):
            for k in range(columnas_matriz1):
                resultado[i][j] += matriz1[i][k] * matriz2[k][j]

    return resultado

def imprimir_matriz(matriz):
    for fila in matriz:
        print(fila)

def main():
    # Ingresar dimensiones de la primera matriz
    filas_matriz1 = int(input("Ingrese el número de filas de la primera matriz: "))
    columnas_matriz1 = int(input("Ingrese el número de columnas de la primera matriz: "))

    # Ingresar la primera matriz
    print("\nIngrese los valores de la primera matriz:")
    matriz1 = ingresar_matriz(filas_matriz1, columnas_matriz1)

    # Ingresar dimensiones de la segunda matriz
    filas_matriz2 = int(input("\nIngrese el número de filas de la segunda matriz: "))
    columnas_matriz2 = int(input("Ingrese el número de columnas de la segunda matriz: "))

    # Validar si el producto es posible
    if columnas_matriz1 != filas_matriz2:
        print("\nError: No se puede multiplicar. El número de columnas de la primera matriz debe ser igual al número de filas de la segunda matriz.")
        return

    # Ingresar la segunda matriz
    print("\nIngrese los valores de la segunda matriz:")
    matriz2 = ingresar_matriz(filas_matriz2, columnas_matriz2)

    # Realizar el producto
    resultado = multiplicar_matrices(matriz1, matriz2)

    # Imprimir el resultado
    print("\nEl resultado del producto es:")
    imprimir_matriz(resultado)

if __name__ == "__main__":
    main()
```

3. Desarrolle un programa que permita obtener la matriz transpuesta de una matriz ingresada. El programa debe validar las condiciones necesarias para ejecutar la operación

```python
def crear_matriz(filas, columnas):
    """
    Función para crear una matriz solicitando los valores al usuario.
    """
    matriz = []  # Inicializamos una lista vacía para almacenar la matriz
    for i in range(filas):  # Iteramos sobre el número de filas
        fila = []  # Inicializamos una lista vacía para cada fila
        for j in range(columnas):  # Iteramos sobre el número de columnas
            # Solicitamos al usuario que ingrese el valor para la posición [i+1][j+1]
            valor = float(input(f"Ingrese el valor para la posición [{i+1}][{j+1}]: "))
            fila.append(valor)  # Agregamos el valor a la fila actual
        matriz.append(fila)  # Agregamos la fila completa a la matriz
    return matriz  # Devolvemos la matriz creada


def imprimir_matriz(matriz):
    """
    Función para imprimir una matriz de forma legible.
    """
    for fila in matriz:  # Iteramos sobre cada fila de la matriz
        print(fila)  # Imprimimos la fila actual


def validar_matriz(matriz):
    """
    Función para validar que la matriz sea válida.
    Una matriz es válida si no está vacía y todas sus filas tienen la misma longitud.
    """
    if not matriz:  # Verificamos si la matriz está vacía
        return False
    longitud_fila = len(matriz[0])  # Longitud de la primera fila
    for fila in matriz:  # Iteramos sobre cada fila
        if len(fila) != longitud_fila:  # Verificamos que todas las filas tengan la misma longitud
            return False
    return True  # La matriz es válida


def obtener_transpuesta(matriz):
    """
    Función para obtener la matriz transpuesta.
    La transpuesta intercambia filas por columnas.
    """
    # Usamos comprensión de listas para crear la matriz transpuesta
    transpuesta = [[matriz[j][i] for j in range(len(matriz))] for i in range(len(matriz[0]))]
    return transpuesta  # Devolvemos la matriz transpuesta


def main():
    """
    Función principal del programa.
    """
    # Solicitar al usuario las dimensiones de la matriz
    filas = int(input("Ingrese el número de filas de la matriz: "))
    columnas = int(input("Ingrese el número de columnas de la matriz: "))

    # Crear la matriz
    print("\nCreando la matriz:")
    matriz = crear_matriz(filas, columnas)

    # Validar que la matriz sea válida
    if not validar_matriz(matriz):
        print("La matriz ingresada no es válida. Asegúrese de que todas las filas tengan la misma longitud.")
        return  # Terminar el programa si la matriz no es válida

    # Mostrar la matriz original
    print("\nMatriz original:")
    imprimir_matriz(matriz)

    # Obtener y mostrar la matriz transpuesta
    transpuesta = obtener_transpuesta(matriz)
    print("\nMatriz transpuesta:")
    imprimir_matriz(transpuesta)


# Punto de entrada del programa
if __name__ == "__main__":
    main()  # Llamar a la función principal para ejecutar el programa
```

4. Desarrollar un programa que sume los elementos de una columna dada de una matriz.

```python
def ingresar_matriz(filas, columnas):
    matriz = []
    for i in range(filas):
        while True:
            fila = input(f"Ingrese la fila {i+1} (separada por espacios): ").strip().split()
            if len(fila) == columnas:
                try:
                    fila = [float(valor) for valor in fila]
                    matriz.append(fila)
                    break
                except ValueError:
                    print("Error: Ingrese solo números válidos.")
            else:
                print(f"Error: La fila debe tener {columnas} elementos.")
    return matriz

def sumar_columna(matriz, columna):
    suma = 0
    for fila in matriz:
        suma += fila[columna]
    return suma

def imprimir_matriz(matriz):
    for fila in matriz:
        print(fila)

def main():
    # Ingresar dimensiones de la matriz
    filas = int(input("Ingrese el número de filas de la matriz: "))
    columnas = int(input("Ingrese el número de columnas de la matriz: "))

    # Ingresar la matriz
    print("\nIngrese los valores de la matriz:")
    matriz = ingresar_matriz(filas, columnas)

    # Mostrar la matriz
    print("\nLa matriz ingresada es:")
    imprimir_matriz(matriz)

    # Seleccionar columna para sumar
    while True:
        columna = int(input(f"\nIngrese el número de la columna a sumar (1 a {columnas}): ")) - 1
        if 0 <= columna < columnas:
            break
        else:
            print(f"Error: La columna debe estar entre 1 y {columnas}.")

    # Calcular y mostrar la suma de la columna
    suma = sumar_columna(matriz, columna)
    print(f"\nLa suma de la columna {columna + 1} es: {suma}")

if __name__ == "__main__":
    main()
```

5. Desarrollar un programa que sume los elementos de una fila dada de una matriz.

```python
def ingresar_matriz(filas, columnas):
    matriz = []
    for i in range(filas):
        while True:
            fila = input(f"Ingrese la fila {i+1} (separada por espacios): ").strip().split()
            if len(fila) == columnas:
                try:
                    fila = [float(valor) for valor in fila]
                    matriz.append(fila)
                    break
                except ValueError:
                    print("Error: Ingrese solo números válidos.")
            else:
                print(f"Error: La fila debe tener {columnas} elementos.")
    return matriz

def sumar_fila(matriz, fila):
    return sum(matriz[fila])

def imprimir_matriz(matriz):
    for fila in matriz:
        print(fila)

def main():
    # Ingresar dimensiones de la matriz
    filas = int(input("Ingrese el número de filas de la matriz: "))
    columnas = int(input("Ingrese el número de columnas de la matriz: "))

    # Ingresar la matriz
    print("\nIngrese los valores de la matriz:")
    matriz = ingresar_matriz(filas, columnas)

    # Mostrar la matriz
    print("\nLa matriz ingresada es:")
    imprimir_matriz(matriz)

    # Seleccionar fila para sumar
    while True:
        fila = int(input(f"\nIngrese el número de la fila a sumar (1 a {filas}): ")) - 1
        if 0 <= fila < filas:
            break
        else:
            print(f"Error: La fila debe estar entre 1 y {filas}.")

    # Calcular y mostrar la suma de la fila
    suma = sumar_fila(matriz, fila)
    print(f"\nLa suma de la fila {fila + 1} es: {suma}")

if __name__ == "__main__":
    main()
```
